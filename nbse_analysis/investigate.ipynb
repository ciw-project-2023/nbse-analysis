```python
from rdkit.Chem import AllChem
from py3Dmol import view
from copy import copy
```

```python
def show_mol(mol):
    buffer = view(width=400, height=400)
    buffer.addModel(AllChem.MolToMolBlock(mol), 'mol')
    buffer.setStyle({'model': -1}, {"stick": {}})
    buffer.zoomTo()
    buffer.show()
```

```python
(mol := AllChem.MolFromSmiles('CCCc1c(Sc2c(Cl)cccc2Cl)[nH]c2nc(N)nc(N)c12'))
```

```python
show_mol(mol)
```

```python
molWithConf = copy(mol)
AllChem.EmbedMolecule(molWithConf, AllChem.srETKDGv3())
```

```python
show_mol(molWithConf)
```

```python
molOptimizedUff = copy(mol)
AllChem.UFFOptimizeMolecule(molOptimizedUff)
show_mol(molOptimizedUff)
```

```python
molOptimizedMmff = copy(mol)
AllChem.MMFFOptimizeMolecule(molOptimizedMmff)
show_mol(molOptimizedMmff)
```

```python
molWithHs = AllChem.AddHs(mol)
```

```python
molWithHsConf = copy(molWithHs)
AllChem.EmbedMolecule(molWithHsConf, AllChem.srETKDGv3())
show_mol(molWithHsConf)
```

```python
molWithHsOptimizedUff = copy(molWithHsConf)
AllChem.UFFOptimizeMolecule(molWithHsOptimizedUff)
show_mol(molWithHsOptimizedUff)
```

```python
molWithHsOptimizedMmff = copy(molWithHsConf)
AllChem.MMFFOptimizeMolecule(molWithHsOptimizedMmff)
show_mol(molWithHsOptimizedMmff)
```
